### int(Integers), Float or Double(decimals), Str(alphabets) Bool(True or False) and None

```python
print(f'2 is of type {type(2)}')
print(f'2.0 is of type {type(2.0)}')
print('2 is different from 2.0 because 2.0 has a decimal')
print(f"'good' is of type {type('good')}")
print(f'True is of type{type(True)}')
```

### List
List contains elements in ordered form. The elements are contained <br>
in a squared bracket. With the elements separated by comma(,)

```python
a = [1,2,3,'a', 'b', True, ['a','b'], ('a',1), {'hi':'greetings'},'etc']
type(a)
```

```python
a = 5 # int
b = 6.0 # float
c = 'Scientific computing' # string
d = ['first_name', 3, True] # list
e = {'program': 'Mathematics'} # dictionaries
f = ('knust', 'legon', 3) # tuple
g = True
h = None
```

```python
ll = [a,b,c,d,e,f,g,h]
for i in ll:
    print(f'{i} is of type {type(i)}\n')
```

\begin{itemize}
    \item A string is a sequence of characters that is in quotation mark and also not separated by ','. eg 'imaging'
    \item Integers are the same as what we know in mathematics. eg 1, 2, 7
    \item A float or floating point number refers to a decimal. eg 1.0, 2.0, 3.1
    \item A bool or boolean refers to valriables that are either True or False
    \item List is container of different data types and structure is [a,b,c,d]
    \item A dictionary is also a container but each item in it has a unique identifier called \textbf{key} eg dict = {key:value}
    \item A tuple is also a container just like a list but it is immutable eg (1,'five')
\end{itemize}

Using the idea of a mathematical structure called set and venn diagram as a basis. You can see that a function is a subset or a subroutine of the main program. From math perspective the main program(MP) should contain all variables in the subroutine but that is not so in programing. lets explain with the second cell below this.

```python
a = "I'm outside printer"
def printer():
    print(a)
    bb = "I'm inside printer and hence i'm in a local scope"
try:
    print(bb)
except NameError:
    print('bb is inside or a local of the function printer and hence' \
         'it cant be accessed outside of the function printer()')
```

In python a function is defined by the keyword 'def'. We will talk more
<br> about a function later on it another notebook. But note that we <br> defined the function \textbf{printer()}  above and you can see that the variable \textbf{'a'}  is global to printer that is why the function could access it and print it but \textbf{'bb'} is local to only the function \textbf{printer()} . So outside the function we can't access \textbf{'bb'} again. So we will get an error

# Data type conversion
data can be converted from one form to another eg
strings can be converted integers or floats etc.

```python
a = '1'
print(type(a))
print("var 'a' is now a string but we converted it to an integer below")
a = int(a)
print(type(a))
print("var 'a' is now converted to an integer with int()")
b = 5.8
print(type(b))
print(f"b={b} is a float")
b = int(b)
print(type(b))
print("b is now converted to a string but notice that the decimal\n place is truncated")
print(f"b={b}")
```
