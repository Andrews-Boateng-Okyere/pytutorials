# FUNCTION, f(x)

\textbf{What} is a \textbf{function}?<br>
In mathematics, a function is a rule or relation that maps a set of objects from one domain to the other. Eg $$f(x) = \frac{1}{2}x + 3$$
<br>
In programming, a function is a block of code or procedure or routine that performs a specific task.

<br>eg
<pre><code>
def f(x):
    y = 0.5*x + 3
    return y
</code></pre>
<br>
From the above definitions and examples you can clearly see that a function in mathematics is similar/equivalent to that in programming.
